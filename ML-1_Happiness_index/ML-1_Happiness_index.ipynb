```python
import pandas as pd
import numpy as np
```

```python
hap_ind = pd.read_csv("World Happiness Report 2022.csv")
```

```python
gdp_cap = pd.read_csv("GDP per capita.csv")
```

```python
gdp_cap.dropna(how="all", inplace=True)
```

```python
hap_ind.columns
```

```python
gdp_cap.columns
```

```python
df1 = hap_ind[['Country', 'Happiness score']]
df1
```

```python
df2 = gdp_cap[['GDP per capita, current prices\n (U.S. dollars per capita)', '2022']]
df2.rename(columns={'GDP per capita, current prices\n (U.S. dollars per capita)':'Country', '2022':'GDP_capita-2022'}, inplace=True)
df2
```

```python
# pd.set_option('display.max_rows', None)
df_final = df1.merge(df2)
df_final
```

```python
df_final.info()
```

```python
df_final['GDP_capita-2022'] = df_final['GDP_capita-2022'].replace('no data', np.nan)
```

```python
df_final
```

```python
df_final.isna().sum()
```

```python
df_final.dropna(inplace=True)
```

```python
df_final.isna().sum()
```

```python
df_final.info()
```

```python
df_final['GDP_capita-2022']= df_final['GDP_capita-2022'].astype('float')
```

```python
df_final.info()
```

```python
df_shuffled = df_final.sample(frac=1)
df_shuffled
```

```python
ratio = 0.8
total_rows = df_shuffled.shape[0]
train_size = int(total_rows*ratio)

train = df_shuffled[0:train_size]
test = df_shuffled[train_size:]
```

```python
display(train, test)
```

```python
x = np.c_[train['GDP_capita-2022']]
y = np.c_[train['Happiness score']]
```

#### Visualization

```python
import matplotlib.pyplot as plt
train.plot.scatter(x='GDP_capita-2022', y='Happiness score')
plt.show()
```

#### Select a linear model

```python
import sklearn.linear_model

model = sklearn.linear_model.LinearRegression()
```

#### Train the model

```python
model.fit(x, y)
```

#### Make model prediction for Cameron

```python
x_new = [[1730.102]] # Cameron GDP per capita
print(model.predict(x_new))
```

```python
test_gdp = np.c_[test['GDP_capita-2022']]
test_predicted = model.predict(test_gdp) # predict on test dataset
```

```python
test_country = np.c_[test['Country']]
final = np.column_stack((test_country, np.c_[test['Happiness score']], test_predicted))
pd.DataFrame(final, columns=['Country', 'Old_Happiness_score', 'Predicted_Happiness_score'])
```
